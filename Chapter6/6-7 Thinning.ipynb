### **图像细化**

* 图像细化是将图像中的线条从多像素宽度减少到单位像素宽度的过程，是模式识别领域重要的处理过程之一，常用在文字识别中，可以有效地将文字细化，增加文字的可辨识度，有效的减少数据量和降低图像的存储难度

* 图像细化一般要求细化后骨架的连通性、原图像的细节特征要较好地保留，线条的端点要保留完好，同时线条交叉点不能发生畸变

* 但并非所有形状的图像都适合进行细化，主要应用在由线条形状组成的问题（例如圆环、文字等），实心圆不适合进行细化

* 目前广泛使用的是并行算法的Zhang方法，用cv.THINNING_ZHANGSUEN表示

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 读取图像并判断是否读取成功
img1 = cv.imread(rootpath +'LearnCV_black.png', cv.IMREAD_GRAYSCALE)
if img1 is None:
    print('Failed to read LearnCV_black.png')
    sys.exit

img2 = cv.imread(rootpath +'OpenCV_4.1.png', cv.IMREAD_GRAYSCALE)
if img2 is None:
    print('Failed to read OpenCV_4.1.png')
    sys.exit
```

```python
# 对图片进行细化
thin1 = cv.ximgproc.thinning(img1, thinningType=0)
thin2 = cv.ximgproc.thinning(img2, thinningType=0)
```

```python
# 展示结果
plt.figure(figsize=(20, 16))
plt.subplot(4, 1, 1)
plt.title('原始图像')
plt.imshow(img1, cmap='gray')
plt.subplot(4, 1, 2)
plt.title('图像细化结果')
plt.imshow(thin1, cmap='gray')
plt.subplot(4, 1, 3)
plt.title('原始图像')
plt.imshow(img2, cmap='gray')
plt.subplot(4, 1, 4)
plt.title('原始图像')
plt.imshow(thin2, cmap='gray')
```
