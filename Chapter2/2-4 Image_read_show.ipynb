### **图像的读取与显示**

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import sys
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
image = cv.imread(rootpath + 'flower.jpg')
# 读取图像，并判断图像是否读取成功
if image is None:
    print('Failed to read flowers.jpg')
    sys.exit
```

#### 图像读取函数

```python
fig, (ax1, ax2,ax3,ax4) = plt.subplots(1,4,figsize=(12,2))

# 图像转换为单通道灰度图像后读取
image_gray = cv.imread(rootpath + 'flower.jpg',cv.IMREAD_GRAYSCALE)
# 图像以原样读取，保留alpha通道(第四个通道)
image_origin = cv.imread(rootpath + 'flower.jpg',cv.IMREAD_UNCHANGED)
# 图像以任何可能的颜色格式读取
image_any = cv.imread(rootpath + 'flower.jpg',cv.IMREAD_ANYCOLOR)
# 保留原图像的深度
image_depth = cv.imread(rootpath + 'flower.jpg',cv.IMREAD_ANYDEPTH)
ax1.imshow(image_gray)
ax2.imshow(image_origin)
ax3.imshow(image_any)
ax4.imshow(image_depth)
plt.show()
```

#### 窗口图像函数与图像显示函数

```python
# 创建窗口，允许用户自由调整窗口大小，调整图像比例以充满窗口，允许添加工具栏和状态栏
window = cv.namedWindow('test', cv.WINDOW_NORMAL | cv.WINDOW_FREERATIO | cv.WINDOW_GUI_EXPANDED | cv.WINDOW_FULLSCREEN)
cv.imshow(window, image)

# 等待按键按下
cv.waitKey(0);
# 结束所有窗口
cv.destroyAllWindows()
```
