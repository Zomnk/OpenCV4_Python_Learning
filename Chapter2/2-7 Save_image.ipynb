### **图像的保存**

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
import matplotlib.pyplot as plt
rootpath = '/home/zomnk/python_ws/02-OpenCV4/my_image/'
```

```python
img = cv.imread('/home/zomnk/python_ws/02-OpenCV4/images/flower.jpg')
# 读取图像，并判断图像是否读取成功
if img is None:
    print('Failed to read flowers.jpg')
    sys.exit
else:
    # 添加alpha通道，alpha为0,完全透明；为255,完全不透明
    zeros = np.ones(img.shape[:2], dtype=img.dtype) * 80
    # img.shape[:2] 获取img的长宽
    print(img.shape[:2], zeros)
    result = cv.merge([img, zeros])
    print('原图的通道数为{}'.format(img.shape[2]))
    print('处理后的通道数为{}'.format(result.shape[2]))

    # 展示图像
    plt.imshow(img)
    plt.show()
    plt.imshow(result)
    plt.show()
    
    # 保存图像
    cv.imwrite(rootpath + 'flower_alpha.png', result)
    
```
