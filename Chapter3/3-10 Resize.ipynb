### **图像尺寸变换**

* INTER_NEAREST 最近邻插值法
* INTER_LINEAR 双线性插值法
* INTER_CUBIC 双三次插值法
* INTER_AREA 使用像素区域关系重新采样，首选用于图像缩小，图像方法时效果与INTER_NEAREST相似
* INTER_LANCZOS4 Lanczos插值法
* INTER_LINEAR_EXACT 位精确双线性插值法
* INTER_MAX 用掩模进行插值

如果要缩小图像，一般用INTER_AREA有较好的效果；放大图像时通常用INTER_CUBIC和INTER_LINEAR有比较好的效果

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 读取图像并判断是否读取成功
img = cv.imread(rootpath +'lena.jpg')[:, :, [2, 1, 0]]
if img is None:
    print('Failed to read lena.jpg')
    sys.exit
```

```python
# 将图像缩小
small_img = cv.resize(img, (15, 15), interpolation=cv.INTER_AREA)
# 最近邻插值法
big_img1 = cv.resize(img, (50, 50), interpolation=cv.INTER_NEAREST)
# 双线性插值法
big_img2 = cv.resize(img, (50, 50), interpolation=cv.INTER_LINEAR)
# 双三次插值法
big_img3 = cv.resize(img, (50, 50), interpolation=cv.INTER_CUBIC)
# Lanczos插值法
big_img4 = cv.resize(img, (50, 50), interpolation=cv.INTER_LANCZOS4)
# 位精确双线性插值法
big_img5 = cv.resize(img, (50, 50), interpolation=cv.INTER_LINEAR_EXACT)
```

```python
# 展示结果
plt.figure(figsize=(18, 6))
plt.subplot(1, 6, 1)
plt.title('small')
plt.imshow(small_img)
plt.subplot(1, 6, 2)
plt.title('最近邻插值法')
plt.imshow(big_img1)
plt.subplot(1, 6, 3)
plt.title('双线性插值法')
plt.imshow(big_img2)
plt.subplot(1, 6, 4)
plt.title('双三次插值法')
plt.imshow(big_img3)
plt.subplot(1, 6, 5)
plt.title('Lanczos插值法')
plt.imshow(big_img4)
plt.subplot(1, 6, 6)
plt.title('精确双线性插值法')
plt.imshow(big_img5)
```
