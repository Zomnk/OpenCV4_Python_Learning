### **图像翻转变换**

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 读取图像并判断是否读取成功
img = cv.imread(rootpath +'lena.jpg')[:, :, [2, 1, 0]]
if img is None:
    print('Failed to read lena.jpg')
    sys.exit
```

```python
# 沿x轴翻转，数值等于0表示沿x轴翻转
img_x = cv.flip(img, 0)
# 沿y轴翻转，数值大于0表示沿y轴翻转
img_y = cv.flip(img, 1)
# 先x轴翻转，再y轴翻转，数值小于0表示绕两条轴旋转
img_xy = cv.flip(img, -1)
```

```python
# 展示结果
plt.figure(figsize=(16, 6))
plt.subplot(1, 4, 1)
plt.title('原图像')
plt.imshow(img)
plt.subplot(1, 4, 2)
plt.title('沿x轴翻转')
plt.imshow(img_x)
plt.subplot(1, 4, 3)
plt.title('沿y轴翻转')
plt.imshow(img_y)
plt.subplot(1, 4, 4)
plt.title('两条轴旋转')
plt.imshow(img_xy)
```
