### **图像多通道分离与合并**

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import sys
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
img = cv.imread(rootpath + 'lena.jpg')[:,:,[2,1,0]]
# 读取图像，并判断图像是否读取成功
if img is None:
    print('Failed to read lena.jpg')
    sys.exit
```

```python
# 通道分离
r, g, b = cv.split(img)
# 创建一个和图像尺寸相同的全0矩阵
zeros = np.zeros(img.shape[:2], dtype='uint8')
```

```python
# 将通道数目相同的图像矩阵合并
bg = cv.merge([b, g, zeros])
gr = cv.merge([zeros, g, r])
br = cv.merge([b, zeros, r])
# 将通道数目不相同的图像矩阵进行合并
# 合成一个6通道的矩阵
bgr_6 = cv.merge([bg, r, zeros, zeros])
```

```python
# 展示结果
plt.figure()
plt.title('Blue')
plt.imshow(b, cmap='gray')
plt.show()
plt.title('Green')
plt.imshow(g, cmap='gray')
plt.show()
plt.title('Red')
plt.imshow(r, cmap='gray')
plt.show()
plt.title('Blue_Green')
plt.imshow(bg)
plt.show()
plt.title('Green_Red')
plt.imshow(gr)
plt.show()
plt.title('Blue_Red')
plt.imshow(br)
plt.show()
```
