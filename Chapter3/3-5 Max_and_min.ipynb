### **两幅图像的比较运算**

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 新建a和b
a = np.array([1, 2, 3.3, 4, 5, 9, 5, 7, 8.2, 9, 10, 2])
b = np.array([1, 2.2, 3, 1, 3, 10, 6, 7, 8, 9.3, 10, 1])
img1 = np.reshape(a, (3, 4))
img2 = np.reshape(b, (3, 4))
img3 = np.reshape(a, (2, 3, 2))
img4 = np.reshape(b, (2, 3, 2))
```

```python
# 对两幅单通道图像进行比较
max12 = cv.max(img1, img2)
min12 = cv.min(img1, img2)
# 对两幅多通道图像进行比较
max34 = cv.max(img3, img4)
min34 = cv.min(img3, img4)
```

```python
# 对两幅彩色图像进行比较
img5 = cv.imread(rootpath + 'lena.jpg')[:,:,[2,1,0]]
img6 = cv.imread(rootpath + 'noobcv.jpg')[:,:,[2,1,0]]
max56 = cv.max(img5, img6)
min56 = cv.min(img5, img6)
plt.figure()
plt.title('conMax')
plt.imshow(max56)
plt.figure()
plt.title('conMin')
plt.imshow(min56)
```

```python
# 对两幅灰度图像进行比较
img7 = cv.cvtColor(img5, cv.COLOR_RGB2GRAY)
img8 = cv.cvtColor(img6, cv.COLOR_RGB2GRAY)
max78 = cv.max(img7, img8)
min78 = cv.min(img7, img8)
plt.figure()
plt.title('conMax_GRAY')
plt.imshow(max78, cmap='gray')
plt.figure()
plt.title('conMin_GRAY')
plt.imshow(min78, cmap='gray')
```

```python
# 与掩模图像进行比较
# 生成一个300*300的低通掩模矩阵
src = np.zeros((512, 512, 3), dtype='uint8')
# 与掩模图像进行比较运算，达到抠图的效果
# 三个通道的300*300都为255
src[100:400:, 100:400:] = 255
min_img5_src = cv.min(img5, src)
plt.figure()
plt.title('Min img5 src')
plt.imshow(min_img5_src)
```

```python
# 生成一个显示红色通道的低通掩模矩阵
src1 = np.zeros((512, 512, 3), dtype='uint8')
# 第1个通道为255, 相当于只保留第1个通道（只显示红色通道）
src1[:, :, 0] = 255
min_img5_src1 = cv.min(img5, src1)
plt.figure()
plt.title('Min img5 src1')
plt.imshow(min_img5_src1)
```
