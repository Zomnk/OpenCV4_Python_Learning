### **感兴趣区域与深浅拷贝**

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 读取图像并判断是否读取成功
img = cv.imread(rootpath +'lena.jpg')[:, :, [2, 1, 0]]
noobcv = cv.imread(rootpath +'noobcv.jpg')[:, :, [2, 1, 0]]
if img is None or noobcv is None:
    print('Failed to read lena.jpg or noobcv.jpg')
    sys.exit
```

```python
mask = cv.resize(noobcv, (200, 200))
# 深拷贝
img1 = img.copy()
# 浅拷贝
img2 = img
# 截取图像的ROI区域
ROI = img[206: 406, 206: 406]
# 深拷贝
ROI_copy = ROI.copy()
# 浅拷贝
ROI1 = ROI
img[206: 406, 206: 406] = mask
# 展示结果
plt.figure(figsize=(8, 8))
plt.subplot(2, 2, 1)
plt.title('img + noobcv1')
plt.imshow(img1)
plt.subplot(2, 2, 2)
plt.title('img + noobcv2')
plt.imshow(img2)
plt.subplot(2, 2, 3)
plt.title('ROI copy1')
plt.imshow(ROI_copy)
plt.subplot(2, 2, 4)
plt.title('ROI copy2')
plt.imshow(ROI1)
```

```python
 # 在图像中绘制圆形
img = cv.circle(img, (300, 300), 20, (0, 0, 255), -1)
# 展示结果
plt.figure(figsize=(8, 8))
plt.subplot(2, 2, 1)
plt.title('img + noobcv1')
plt.imshow(img1)
plt.subplot(2, 2, 2)
plt.title('img + noobcv2')
plt.imshow(img2)
plt.subplot(2, 2, 3)
plt.title('ROI copy1')
plt.imshow(ROI_copy)
plt.subplot(2, 2, 4)
plt.title('ROI copy2')
plt.imshow(ROI1)
```
