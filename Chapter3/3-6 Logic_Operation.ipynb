### **两幅图像的逻辑运算**

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib.pyplot as plt

rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 创建两幅黑白图像
img1 = np.zeros((200, 200), dtype='uint8')
img2 = np.zeros((200, 200), dtype='uint8')
img1[50:150, 50:150] = 255
img2[100:200 ,100:200] = 255
```

```python
# 读取图像并判断是否读取成功
img = cv.imread(rootpath +'lena.jpg')[:,:,[2,1,0]]
if img is None:
    print('Failed to read lena.jpg')
    sys.exit
```

```python
# 进行逻辑运算
Not = cv.bitwise_not(img1)
And = cv.bitwise_and(img1, img2)
Or = cv.bitwise_or(img1, img2)
Xor = cv.bitwise_xor(img1, img2)
img_Not = cv.bitwise_not(img)
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(2, 4, 1)
plt.title('img1')
plt.imshow(img1)
plt.subplot(2, 4, 2)
plt.title('img2')
plt.imshow(img2)
plt.subplot(2, 4, 3)
plt.title('Not')
plt.imshow(Not)
plt.subplot(2, 4, 4)
plt.title('And')
plt.imshow(And)
plt.subplot(2, 4, 5)
plt.title('Or')
plt.imshow(Or)
plt.subplot(2, 4, 6)
plt.title('Xor')
plt.imshow(Xor)
plt.subplot(2, 4, 7)
plt.title('img')
plt.imshow(img)
plt.subplot(2, 4, 8)
plt.title('img_Not')
plt.imshow(img_Not)
```
