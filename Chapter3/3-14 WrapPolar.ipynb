### **图像极坐标变换**

warpPolar()函数中flags参数可选择的标志：
* WARP_POLAR_LINEAR 极坐标变换
* WARP_POLAR_LOG 半对数极坐标变换
* WARP_INVERSE_MAP 逆变换

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 读取图像并判断是否读取成功
img = cv.imread(rootpath +'dial.png')[:, :, [2, 1, 0]]
if img is None:
    print('Failed to read dial.png')
    sys.exit
```

```python
# 获取图像的高和宽
h, w = img.shape[:-1]
# 计算极坐标在图像中的原点
center = (w / 2.0, h / 2.0)
# 极坐标正变换
img_res = cv.warpPolar(img, (300, 600), center, center[0], cv.INTER_LINEAR + cv.WARP_POLAR_LINEAR)
# 极坐标正变换
img_res1 = cv.warpPolar(img_res, (w, h), center, center[0], cv.INTER_LINEAR + cv.WARP_INVERSE_MAP)
# 半对数极坐标变换
img_res2 = cv.warpPolar(img, (300, 600), center, center[0], cv.INTER_LINEAR + cv.WARP_POLAR_LOG)
```

```python
# 展示结果
plt.figure(figsize=(18, 5))
plt.subplot(1, 4, 1)
plt.title('原图像')
plt.imshow(img)
plt.subplot(1, 4, 2)
plt.title('极坐标正变换')
plt.imshow(img_res)
plt.subplot(1, 4, 3)
plt.title('极坐标逆变换')
plt.imshow(img_res1)
plt.subplot(1, 4, 4)
plt.title('半对数极坐标变换')
plt.imshow(img_res2)
```
