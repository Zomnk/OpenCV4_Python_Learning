### **绘制几何图形和生成文字**

lineType的参数有，从上到下质量上升，速度下降：
* cv.LINE_4
* cv.LINE_8
* cv.LINE_AA
* cv.FILLED

字体类型的参数有：
* cv2.FONT_HERSHEY_SIMPLEX 正常大小的无衬线字体
* cv2.FONT_HERSHEY_PLAIN 小尺寸的无衬线字体
* cv2.FONT_HERSHEY_DUPLEX 正常大小的较复杂的无衬线字体
* cv2.FONT_HERSHEY_COMPLEX 正常大小的衬线字体
* cv2.FONT_HERSHEY_TRIPLEX 正常大小的较复杂衬线字体
* cv2.FONT_HERSHEY_COMPLEX_SMALL 小尺寸的衬线字体
* cv2.FONT_HERSHEY_SCRIPT_SIMPLEX 手写风格的字体
* cv2.FONT_HERSHEY_SCRIPT_COMPLEX 复杂的手写风格字体
* cv2.FONT_ITALIC 斜体字体

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 生成一幅黑色图像，用以在其中绘制图形
img = np.zeros((512, 512, 3), dtype='uint8')
# 开始绘制图形
# 绘制实心圆
img = cv.circle(img, (50, 50), 25, (255, 255, 255), -1)
# 绘制空心圆
img = cv.circle(img, (100, 50), 25, (255, 255, 255), 4)
# 绘制直线
img = cv.line(img, (100, 100), (200, 100), (255, 255, 255), 2, cv.LINE_4, 0)
# 绘制椭圆
img = cv.ellipse(img, (300, 255), (100, 70), 0, 0, 270, (255, 255, 255), -1)
# 用一个点近似一个椭圆(
points = cv.ellipse2Poly((200, 400), (100, 70), 0, 0 ,360, 2)
# 使用直线将上述点显示出来
for i in range(len(points) - 1):
    img = cv.line(img, (points[i][0], points[i][1]), (points[i+1][0], points[i+1][1]), (255, 0, 0), 2, cv.LINE_4, 0)
img = cv.line(img, (points[-1][0], points[-1][1]), (points[0][0], points[0][1]), (255, 0, 0), 2, cv.LINE_4, 0)
# 绘制矩形
img = cv.rectangle(img, (50, 400), (100, 450), (0, 255, 0), -1)
img = cv.rectangle(img, (400, 450, 60, 50), (0, 0, 255), 2)
# 绘制多边形
pts = np.array([[350, 83], [463, 90], [500, 171], [421, 194], [338, 141]], dtype='int32')
img = cv.fillPoly(img, [pts], (255, 255, 0), 8)
# 添加文字
img = cv.putText(img, 'Learn OpenCV', (150, 70), 2, 1, (0, 255, 0))
# 展示结果
plt.figure(figsize=(5, 5))
plt.imshow(img)
```
