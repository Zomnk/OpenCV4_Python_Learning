### **图像连接**

* vconcat() 要求输入图像具有相同的长度、数据类型和通道数
* hconcat() 要求输入图像具有相同的高度、数据类型和通道数

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib.pyplot as plt

rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 矩阵的垂直和水平连接
# 定义矩阵A和矩阵B
A = np.array([[1, 7], [2, 8]])
B = np.array([[4, 10],[5, 11]])
# 垂直连接
V_C = cv.vconcat((A, B))
# 水平连接
H_C = cv.hconcat((A, B))
print('垂直连接结果: \n{}'.format(V_C))
print('水平连接结果: \n{}'.format(H_C))
```

```python
# 图像的垂直和水平连接
# 读取四张图像
# 读取图像并判断是否读取成功
img00 = cv.imread(rootpath +'lena00.jpg')[:, :, [2, 1, 0]]
img01 = cv.imread(rootpath +'lena01.jpg')[:, :, [2, 1, 0]]
img10 = cv.imread(rootpath +'lena10.jpg')[:, :, [2, 1, 0]]
img11 = cv.imread(rootpath +'lena11.jpg')[:, :, [2, 1, 0]]
if img00 is None or img01 is None or img10 is None or img11 is None:
    print('Failed to read lena.jpg')
    sys.exit
```

```python
# 图像连接
# 水平连接
img0 = cv.hconcat((img00, img01))
img1 = cv.hconcat((img10, img11))
# 垂直连接
img = cv.vconcat((img0, img1))
```

```python
# 显示结果
plt.figure(figsize=(14, 6))
plt.subplot(2, 4, 1)
plt.title('img00')
plt.imshow(img00)
plt.subplot(2, 4, 2)
plt.title('img01')
plt.imshow(img01)
plt.subplot(2, 4, 3)
plt.title('img0')
plt.imshow(img0)
plt.subplot(2, 4, 5)
plt.title('img10')
plt.imshow(img10)
plt.subplot(2, 4, 6)
plt.title('img11')
plt.imshow(img11)
plt.subplot(2, 4, 7)
plt.title('img1')
plt.imshow(img1)
plt.subplot(1, 4, 4)
plt.title('img')
plt.imshow(img)
```
