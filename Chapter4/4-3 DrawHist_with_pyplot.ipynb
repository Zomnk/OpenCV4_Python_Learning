### **用matplotliub绘制灰度直方图**

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 读取图像并判断是否读取成功
img = cv.imread(rootpath +'flower.jpg')
if img is None:
    print('Failed to read flower.jpg')
    sys.exit
```

```python
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title('原始图像')
plt.imshow(img[:, :, [2, 1, 0]])
plt.subplot(1, 2, 2)
plt.title('灰度直方图')
# img.ravel()将多维数据展平为一维数据
_, _, _ = plt.hist(x=img.ravel(), bins=256, range=[0, 256])
plt.show()
```
