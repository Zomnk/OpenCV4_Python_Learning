### **直方图的计算**

直方图就是统计图像中每个灰度值的个数，以灰度值的个数或比重值作为纵轴绘制的统计图，灰度值代表亮暗程度，可以根据直方图分析图像的亮暗对比度，并调整图像的亮暗程度

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 不适用科学计数法，完整显示数字
np.set_printoptions(suppress=True)
```

```python
# 读取图像并判断是否读取成功
img = cv.imread(rootpath +'apple.jpg')
if img is None:
    print('Failed to read apple.jpg')
    sys.exit
```

```python
# 对图像进行直方图计算
# 对于灰度图，第二个参数的尺寸为[0]
# 对于彩色图，第二个参数尺寸为[0], [1], [2]，分别对应B、G、R三个通道
hist = cv.calcHist([img], [0], None, [256], [0, 256])
```

```python
# 输出结果
print('统计灰度的直方图为\n{}'.format(hist))
```
