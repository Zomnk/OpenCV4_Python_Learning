### **直方图均衡化**

如果一幅图像的直方图集中在一个区域，对比度低，呈现一种模糊的感觉，映射将图像中灰度值的范围扩大，增加原来两个灰度值之间的差值，可以将图像中的纹理凸显出来，称为直方图的均衡化

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 读取图像并判断是否读取成功
img = cv.imread(rootpath +'equalizeHist.jpg', 0)
if img is None:
    print('Failed to read equalizeHist.jpg')
    sys.exit
```

```python
# 进行直方图的均衡化
img_result = cv.equalizeHist(img) # 必须是八位单通道图像
# 绘制均衡化前后的直方图
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title('原始图像')
plt.hist(img.ravel(), 256, [0, 256])
plt.subplot(1, 2, 2)
plt.title('均衡化后的图像')
plt.hist(img_result.ravel(), 256, [0, 256])
```

```python
# 展示直方图均衡化前后的图像
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title('原始图像')
plt.imshow(img, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('均衡化后的图像')
plt.imshow(img_result, cmap='gray')
```
