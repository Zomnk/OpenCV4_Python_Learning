### **用matplotliub绘制BGR通道直方图**

先用cv.calcHist计算每个通道的直方图，再用plot函数实现绘制

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 读取图像并判断是否读取成功
img = cv.imread(rootpath +'flower.jpg')
if img is None:
    print('Failed to read flower.jpg')
    sys.exit
```

```python
# 绘制直方图并展示
color = ('b', 'g', 'r')
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title('原始图像')
plt.imshow(img[:, :, [2, 1, 0]])
plt.subplot(1, 2, 2)
plt.title('灰度直方图')
for i, col in enumerate(color):
    hist_item = cv.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist_item, color=col)
plt.show()
```
