### **中值滤波**

非线性滤波的计算过程包含排序、逻辑计算等，线性滤波只能使噪声变得更柔和，而非线性滤波一定程度上可以去除噪声

中值滤波取滤波器范围内所有像素值的中值替代滤波器中心的像素值，能有效抑制噪声的非线性信号，不依赖滤波器内与典型值相差较大的值，因此对斑点噪声和椒盐噪声具有很好的处理效果

相比于均值滤波，中值滤波对脉冲干扰信号的处理效果更佳，在一定程度下对图像的边缘信息保护得更好，可以避免细节的模糊，但耗时更长

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 读取图像并判断是否读取成功
gray = cv.imread(rootpath +'GraySaltPepperImage.jpg')
if gray is None:
    print('Failed to read GraySaltPepperImage.jpg')
    sys.exit

img = cv.imread(rootpath +'ColorSaltPepperImage.jpg')[:,:,[2,1,0]]
if img is None:
    print('Failed to read ColorSaltPepperImage.jpg')
    sys.exit
```

```python
# 分别对含椒盐噪声的彩色图像和灰度图像进行中值滤波
img_3 = cv.medianBlur(img, 3)
gray_3 = cv.medianBlur(gray, 3)
```

```python
# 增大ksize，图像会变模糊
img_9 = cv.medianBlur(img, 9)
gray_9 = cv.medianBlur(gray, 9)
```

```python
# 展示结果
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('原灰度图像')
plt.imshow(gray, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('3×3中值滤波')
plt.imshow(gray_3, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('9×9中值滤波')
plt.imshow(gray_9, cmap='gray')
```

```python
# 展示结果
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('原彩色图像')
plt.imshow(img)
plt.subplot(1, 3, 2)
plt.title('3×3中值滤波')
plt.imshow(img_3)
plt.subplot(1, 3, 3)
plt.title('9×9中值滤波')
plt.imshow(img_9)
```
