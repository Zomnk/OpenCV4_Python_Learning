### **均值滤波**

均值滤波将滤波器内所有像素值都看做中心像素值的测量值，取平均值作为滤波器中心的像素值，滤波器内每个数据表示对应像素在决定中心像素值的过程中所占权重

均值滤波的优点是在像素值变化趋势一致的情况下，可以将受噪声影响而突然变化的像素值修正为周围邻近像素的平均值，但是会缩小像素值之间的差距，使得细节信息变得模糊，滤波器范围越大，变模糊的效果越显著

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
def my_blur(image):
    return cv.blur(image, (3, 3)), cv.blur(image, (9, 9))
```

```python
# 读取图像并判断是否读取成功
img = cv.imread(rootpath +'Gray_dolphins.jpg')
if img is None:
    print('Failed to read Gray_dolphins.jpg')
    sys.exit

img_sp = cv.imread(rootpath +'GraySaltPepperImage.jpg')
if img_sp is None:
    print('Failed to read GraySaltPepperImage.jpg')
    sys.exit

img_gauss = cv.imread(rootpath +'GrayGaussImage.jpg')
if img_gauss is None:
    print('Failed to read GrayGaussImage.jpg')
    sys.exit
```

```python
# 进行均值滤波
img1, img2 = my_blur(img)
img_sp1, img_sp2 = my_blur(img_sp)
img_gauss1, img_gauss2 = my_blur(img_gauss)
```

```python
# 展示结果
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('原灰度图像')
plt.imshow(img, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('3×3均值滤波')
plt.imshow(img1, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('9×9均值滤波')
plt.imshow(img2, cmap='gray')
```

```python
# 展示结果
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('原椒盐噪声图像')
plt.imshow(img_sp, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('3×3均值滤波')
plt.imshow(img_sp1, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('9×9均值滤波')
plt.imshow(img_sp2, cmap='gray')
```

```python
# 展示结果
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('原高斯噪声图像')
plt.imshow(img_gauss, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('3×3均值滤波')
plt.imshow(img_gauss1, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('9×9均值滤波')
plt.imshow(img_gauss2, cmap='gray')
```

可以发现，滤波器越大，滤波后图像变得越模糊
