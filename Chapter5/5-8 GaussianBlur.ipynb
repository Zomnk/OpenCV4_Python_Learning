### **高斯滤波**

高斯滤波器考虑像素与滤波器中心距离的影响，以滤波器中心为高斯分布的均值

```python
# -*- coding:utf-8 -*-
import cv2 as cv
import numpy as np
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# 设置中文字体
matplotlib.rcParams['font.family'] = 'simhei'
matplotlib.rcParams['axes.unicode_minus'] = False
rootpath = '/home/zomnk/python_ws/02-OpenCV4/images/'
```

```python
# 读取图像并判断是否读取成功
img = cv.imread(rootpath +'Gray_dolphins.jpg')
if img is None:
    print('Failed to read Gray_dolphins.jpg')
    sys.exit

img_sp = cv.imread(rootpath +'GraySaltPepperImage.jpg')
if img_sp is None:
    print('Failed to read GraySaltPepperImage.jpg')
    sys.exit

img_gauss = cv.imread(rootpath +'GrayGaussImage.jpg')
if img_gauss is None:
    print('Failed to read GrayGaussImage.jpg')
    sys.exit
```

```python
# 分别对上述图像进行高斯滤波，x方向上标准差为10，y方向上标准差为20
result_5 = cv.GaussianBlur(img, (5, 5), 10, 20)
result_9 = cv.GaussianBlur(img, (9, 9), 10, 20)
result_5_salt = cv.GaussianBlur(img_sp, (5, 5), 10, 20)
result_9_salt = cv.GaussianBlur(img_sp, (9, 9), 10, 20)
result_5_gauss = cv.GaussianBlur(img_gauss, (5, 5), 10, 20)
result_9_gauss = cv.GaussianBlur(img_gauss, (9, 9), 10, 20)
```

```python
# 展示结果
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('原灰度图像')
plt.imshow(img, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('5×5高斯滤波')
plt.imshow(result_5, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('9×9均值滤波')
plt.imshow(result_9, cmap='gray')
```

```python
# 展示结果
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('椒盐噪声灰度图像')
plt.imshow(img_sp, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('5×5高斯滤波')
plt.imshow(result_5_salt, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('9×9均值滤波')
plt.imshow(result_9_salt, cmap='gray')
```

```python
# 展示结果
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('高斯噪声灰度图像')
plt.imshow(img_gauss, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('5×5高斯滤波')
plt.imshow(result_5_gauss, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('9×9均值滤波')
plt.imshow(result_9_gauss, cmap='gray')
```

可以发现高斯滤波对高斯噪声去除效果较好，但是同样会造成模糊，滤波器尺寸越大越模糊
