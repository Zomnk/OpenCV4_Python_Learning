### **图像滤波**

* 图像滤波是指去除图像中不重要的内容而使关心的内容表达的更加清晰的方法，根据目的，图像滤波分为消除图像噪声的滤波和提取图像中部分特征信息的滤波

* 去除图像噪声叫做图像的平滑或图像去噪，由于噪声信号在图像中主要集中在高频段，因此使用的主要是低通滤波器

* 图像中纹理变化比较明显的区域信号频率很高，因此使用高通滤波器对图像进行处理可以起到提取、增强图像边缘信息和锐化图像的作用

* 常用图像模糊来代替图像的低通滤波，因为图像低通滤波在去除图像噪声的同时会将图像的边缘信息弱化，使得整幅图像变得模糊，模糊和滤波是等价的

* 用“滤波”指图像的“低通滤波”，用“边缘检测”指图像的“高通滤波”

* 图像滤波分为线性滤波和非线性滤波，线性滤波包括均值滤波、方框滤波和高斯滤波；常见的非线性滤波宝库叫哦中值滤波和双边滤波

* 图像的滤波操作和图像的卷积操作类似，不同之处在于图像滤波不需要将滤波模板旋转180°，卷积模板在图像滤波中称为滤波模板、滤波器或邻近算子，通过滤波范围内所有像素值之间的线性组合可以求取中心像素滤波后的像素值，因此这种方式称为线性滤波
